### a. histogram of the image

### b. Convert image into gray scale and obtain histogram of the image


```python
gray_scale_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

histr1 = cv2.calcHist([gray_scale_image],[0],None,[256],[0,256]) 
  
# show the plotting graph of an image 
plt.plot(histr1) 
cv2.imshow('Grayscale Image',gray_scale_image)
cv2.waitKey(0)
plt.show() 
```

### c. Convert image into binary and obtain histogram of the image


```python
(thresh, blackAndWhiteImage) = cv2.threshold(gray_scale_image, 120,255, cv2.THRESH_BINARY)

histr2 = cv2.calcHist([blackAndWhiteImage],[0],None,[256],[0,256]) 
  
# show the plotting graph of an image 
plt.plot(histr2) 
cv2.imshow('Binary Image',blackAndWhiteImage)
cv2.waitKey(0)
plt.show() 
```
